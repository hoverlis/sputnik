```python
from osgeo import gdal
import pandas as pd
import os
```

```python
df = pd.DataFrame(columns=['Band', 'Wavelength'])
```

```python
def get_wavelengths(bands_list):
  wavelengths = []
  for n in bands_list:
    if 'LC09' or 'LC08' in n and "xml" not in n:
      if "B2" in n:
        wavelengths.append(0.48)
      elif "B3" in n:
        wavelengths.append(0.56)
      elif "B4" in n:
        wavelengths.append(0.65)
      elif "B5" in n:
        wavelengths.append(0.86)
      elif "B6" in n:
        wavelengths.append(1.61)
      elif "B7" in n:
        wavelengths.append(2.20)
      elif "B10" in n:
        wavelengths.append(10.89)
      elif "B11" in n:
        wavelengths.append(12.00)
      else:
        wavelengths.append(0.44)
  return wavelengths

# Расписать для 8
```

```python
dir = "/content/drive/MyDrive/Project_BV/Clip/clip_Hvostohranilish/clip_Kachkanar/"
band_paths = []
mean_vals = []
dic = {}
for b in os.listdir(dir):
  if 'xml' not in b:
    band_paths.append(dir+b)
for b in band_paths:
  raster = gdal.Open(b)
  band = raster.GetRasterBand(1)
  mean = band.ComputeStatistics(0)[2]*0.0001
  mean_vals.append(mean)
  dic[b] = mean
df['Band'] = sorted(dic.keys())
df['Wavelength'] = get_wavelengths(sorted(band_paths))
df['Kachkanar'] = [dic[name] for name in sorted(dic.keys())]
df
```
